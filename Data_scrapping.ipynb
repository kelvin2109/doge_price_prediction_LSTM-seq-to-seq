```python
def convert_into_df(response):
    
    aList = json.loads(response.text)
    doge_df = pd.DataFrame(aList)
    return doge_df
```

```python
def convert_date_and_rename_features(df):
    for i in range(len(df)):
        df['time_close'][i] = df['time_close'][i].split("T",1)[0]
        
    df.rename(columns={'time_close':'Date','rate_open':'Open','rate_high':'High','rate_low':'Low','rate_close':'Close'},inplace=True)
    df.index = df['Date']
    df.drop(columns=['Date'],inplace=True)
    return df
```
